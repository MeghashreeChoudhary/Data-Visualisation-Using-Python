# Question 1
  What is Plotly?

1. Plotly is a platform to create data visualization. 
2. Plotly's Python graphing library makes interactive, publication-quality graphs.
3. This library supports over 40 unique chart types covering a wide range of statistical, financial, geographic, scientific, and    3-dimensional use-cases.
4. The graphs and charts can be displayed in Jupyter notebooks, saved to standalone HTML files, or served as part of pure Python-built web applications using Dash.
5. You can use it with python, R, Julia, and a few other languages.

Where is Plotly used?

1. Main benefit of using Plotly on Python (with pandas using cufflinks) is that it allows you to create interactive charts/dashboards compared to the static ones from matplotlib/seaborn.
2. Figure Converters which convert matplotlib, ggplot2, and IGOR Pro graphs into interactive, online graphs.
3. Plotly Apps for Google Chrome.
4. Plotly.js is an open source JavaScript library for creating graphs and dashboards.


# Question 3
Plot a Pie Chart of Indian Car companies market share % values (max 6 Companies market share).


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
share = [50, 17.4, 5.8, 5.6, 3.3, 3.1]
companies = ["Maruti Suzuki", "Hyundai", "Tata", "Kia", "Mahindra", "Renault"]
data = {"company_name": companies,"market_share": share}
df = pd.DataFrame(data)
plt.pie(share, labels = companies)
plt.show()
```

# Question 4
Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )


```python
active_cases = [117691, 105662, 34257, 31218, 26710, 21631, 20170, 14191, 7706, 5063]
states = ["Kerala", "Maharashtra", "Karnataka", "Tamil Nadu", "Andhra Pradesh", "Assam", "West Bengal", "Telangana", "Manipur", "Mizoram"]
data = {"active_cases": active_cases,"states": states}
df = pd.DataFrame(data)
plt.figure(figsize=(15, 6))
plt.title("Covid-19 cases in India")
plt.bar(states, active_cases)
plt.show()
```
